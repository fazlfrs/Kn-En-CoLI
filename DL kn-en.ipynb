```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
import warnings
warnings.filterwarnings(action='ignore')
import pandas as pd
import gensim
from gensim.models import Word2Vec
```

```python
df=open('rawTexts/rawtext.txt',encoding='utf-8').read().split('\n')
data=[]
for i in df:
    temp=[]
    for j in word_tokenize(i):
        temp.append(j.lower())
    data.append(temp)
```

```python
#using skip gram
w2v = gensim.models.Word2Vec(data, min_count = 1, size = 200, window = 5, sg = 1) 
print("'likes' " + "and 'views' - Skip Gram : ", w2v.similarity('likes', 'views'))
```

```python
vocab=list(w2v.wv.vocab)

chars=[]
for word in vocab:
    temp=[]
    for char in word:
        temp.append(char)
    chars.append(temp)
```

```python
c2v = gensim.models.Word2Vec(chars, min_count = 1, size = 30, window = 5, sg = 1) 
```

```python
#sub word embedding copy
#!pip install BPEmb
from bpemb import BPEmb
bpemb_en = BPEmb(lang="en",vs=50000)
```

```python
data=[]
for i in df:
    
    for j in word_tokenize(i):
        temp=[]
        for sw in bpemb_en.encode(j):
            if sw[0]=='▁':sw=sw[1:]
            temp.append(sw.lower())
        data.append(temp)
```

```python
sw2v=gensim.models.Word2Vec(data, min_count = 1, size = 100, window = 5, sg = 1)
```

```python
import pickle
import numpy as np

embeddings_index = {}
mx=700
mx1=1000
words1=[]
for line in df:
    words = word_tokenize(line)
    for word in words:
        word=word.lower()
        words1.append(words)
        vec=[]
        vec.extend(w2v[word])
        for sw in bpemb_en.encode(word):
            if sw.startswith('▁'):sw=sw[1:]
            vec.extend(sw2v[sw.lower()])
        if len(vec)>mx : continue
        elif len(vec)<mx: vec.extend([0.0] * (mx - len(vec)))

        for char in word:
            vec.extend(c2v[char])
        if len(vec)>mx1 : continue
        elif len(vec)<mx1: vec.extend([0.0] * (mx1 - len(vec)))

        embeddings_index[word] =vec
```

```python
df['word']=[str(w).lower() for w in df.word]

words=list(set(df.word.values))
words.append("ENDPAD")
n_words = len(words)
```

```python
tags=[]
for t in df.tag:
    t=t.strip()
    if t=='em':t='en'
    if t=='uni':t='univ'
    if t=='k':t='kn'
    tags.append(t)
df.tag=tags
tags = list(set(df["tag"].values))
n_tags = len(tags)
```

```python
word2int = {}
int2word = {}

for i, word in enumerate(words):
	word2int[word] = i+1
	int2word[i+1] = word

tag2int = {}
int2tag = {}
```

```python

#Extracting Sentences from dataset
sentences=[]
sent=[]
for i in range(len(df.word)):
    if(df['word'][i]!='.'):
        sent.append([str(df.word[i]).lower(),df['tag'][i]])
    else:
        sent.append(['.','univ'])
        sentences.append(sent)
        sent=[]
```

```python
for i, tag in enumerate(tags):
	tag2int[tag] = i+1
	int2tag[i+1] = tag

X = [[word2int[w[0]] for w in s] for s in sentences]
y = [[tag2int[w[1]] for w in s] for s in sentences]

print('sample X_numberised: ',X[42],'\n')
print('sample y_numberised: ',y[42],'\n')
```

```python
pckl_files=[X,y,word2int,int2word,tag2int,int2tag]

with open('ka-en-data.pkl','wb') as f:
	pickle.dump(pckl_files,f)
print('Data saved as pickle file')
```

```python
def load_data(data):
	dpth=data
#	mpth=w2v
	with open(dpth, 'rb') as f:
		files=pickle.load(f)
		#X, Y, word2int, int2word, tag2int, int2tag = pickle.load(f)
#	with open(mpth, 'rb') as g:
#		embeddings_index = pickle.load(g)
	return files#,embeddings_index
```

```python
files=load_data('ka-en-data1.pkl')
X, Y, word2int, int2word, tag2int, int2tag=files[0],files[1],files[2],files[3],files[4],files[5]

    
X = pad_sequences(X, maxlen=MAX_SEQUENCE_LENGTH)
Y = pad_sequences(Y, maxlen=MAX_SEQUENCE_LENGTH)

print(tag2int)
n_tags = len(tag2int)
```

```python

	Y = to_categorical(Y, num_classes= len(tag2int) + 1)

	print('TOTAL TAGS', len(tag2int))
	print('TOTAL WORDS', len(word2int))
```

```python
	indices = np.arange(X.shape[0])
	np.random.shuffle(indices)

	X = X[indices]
	Y = Y[indices]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
x_train=np.array(x_train)
```

```python
	print('Total %s word vectors.' % len(embeddings_index))
	embedding_matrix = np.random.random((len(word2int) + 1, EMBEDDING_DIM))
```

```python
embedding_matrix.shape
```

```python
	for word, i in word2int.items():
    		embedding_vector = embeddings_index.get(word)
    		if embedding_vector is not None:
        		embedding_matrix[i] = embedding_vector

	print('Embedding matrix shape', np.array(embedding_matrix).shape)
	print('x_train shape', x_train.shape)
```

```python
	embedding_layer = Embedding(len(word2int) + 1,
                          	  EMBEDDING_DIM,
                          	  weights=[embedding_matrix],
                          	  input_length=MAX_SEQUENCE_LENGTH,
                         	   trainable=True)
	sequence_input = Input(shape=(MAX_SEQUENCE_LENGTH,), dtype='int32')
	embedded_sequences = embedding_layer(sequence_input)
```

```python
	l_lstm = Bidirectional(LSTM(69, return_sequences=True))(embedded_sequences)
	preds = TimeDistributed(Dense(n_tags + 1, activation='softmax'))(l_lstm)
	model = Model(sequence_input, preds)
```

```python
	model.compile(loss='categorical_crossentropy',optimizer='rmsprop',metrics=['acc'])
	
	print("model fitting - Bidirectional LSTM")
	model.summary()
```

# after 20 epochs

```python
test_pred = model.predict(x_train, verbose=1)

idx2tag = {i: w for w, i in tag2int.items()}

def pred2label(pred):
    out = []
    for pred_i in pred:
        
        out_i = []
        for p in pred_i:
            tmp=np.argmax(p)
            if tmp==0 :tmp=2
            p_i = tmp
            
            out_i.append(idx2tag[p_i].replace("PAD", "univ"))
            
        out.append(out_i)
    return out

pred_labels = pred2label(test_pred)
test_labels = pred2label(y_train)

print('accuracy: ',accuracy_score(test_labels, pred_labels))
print('f1 score: ',f1_score(test_labels, pred_labels))
print('recall_score: ',recall_score(test_labels, pred_labels))
print('precision_score: ',precision_score(test_labels, pred_labels))

print(classification_report(test_labels, pred_labels))
```
